```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler


X = pd.read_csv('arcene_train.data', delim_whitespace=True, header=None)
y = pd.read_csv('arcene_train.labels', header=None)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


pca = PCA(n_components=3)
X_pca = pca.fit_transform(X_scaled)

df_pca = pd.DataFrame(X_pca)
df_pca['label'] = y


df_pca.to_csv('arcene_pca_with_labels.csv', index=False)

print('PCA completed. Data saved as arcene_pca_with_labels.csv')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=42)

clf = SVC()
clf.fit(X_train, y_train.values.ravel())

y_pred = clf.predict(X_test)

print('Accuracy:', accuracy_score(y_test, y_pred))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

scatter = ax.scatter(
    X_pca[:, 0], X_pca[:, 1], X_pca[:, 2],
    c=y, cmap='coolwarm', edgecolor='k', alpha=0.7
)

ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')
ax.set_title('3D PCA Visualization of Arcene Dataset')
legend = ax.legend(*scatter.legend_elements(), title="Labels")
ax.add_artist(legend)
plt.show()
```
